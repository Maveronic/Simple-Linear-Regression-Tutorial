```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('realestate.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['price']
x1 = data['size']
```

```python
plt.scatter(x1,y)
yhat = (1.019 * 10**5) + 223.1787*x1
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
fig = plt.plot(x1, yhat, label='linear regression', color='orange')
plt.legend()
plt.title('Linear regression for the price of a houses in an estate')
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
